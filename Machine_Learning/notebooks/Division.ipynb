```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Cargar el archivo CSV
file_path = 'Archivos_CSV_con_etiquetas/Mix_T337+Erika_Completo.csv'
df = pd.read_csv(file_path)

# Dividir el DataFrame en dos partes con la misma cantidad de datos de cada actividad
df_train, df_test = train_test_split(df, test_size=0.5, stratify=df['Actividades'])

# Ordenar los DataFrames resultantes por el Ã­ndice original
df_train = df_train.sort_index()
df_test = df_test.sort_index()

# Guardar los DataFrames resultantes en nuevos archivos CSV
df_train.to_csv('Archivos_CSV_con_etiquetas/Mix_T337+Erika_Completo_train.csv', index=False)
df_test.to_csv('Archivos_CSV_con_etiquetas/Mix_T337+Erika_Completo_test.csv', index=False)
```
